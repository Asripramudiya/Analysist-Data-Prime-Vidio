```python
import numpy as np
import pandas as pd
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

The data variables that selected for analysis are categories, country, show type, director, and ratings.

```python
df = pd.read_csv('amazon_prime_titles.csv', header=0)
df
```

```python
df.rename(columns={'listed_in':'categories'}, inplace=True)
df
```

```python
df.info()
```

```python
df.drop(['show_id'], axis=1, inplace=True)
df.drop(['description'], axis=1, inplace=True)
```

```python
df
```

```python
df.shape
```

```python
df.info()
```

```python
#Menangani nilai duplikat
df.duplicated().sum() 
```

```python
#Menangani missing values
df.isnull().sum()
```

```python
df['director'].fillna('Unavailable', inplace=True)
df['cast'].fillna('Unavailable', inplace=True)
df['country'].fillna('Unavailable', inplace=True)
df['date_added']=df['date_added'].ffill()
df['rating']=df['rating'].fillna(df['rating'].mode()[0])
df
```

```python
df['date_added']=pd.to_datetime(df['date_added'],format='%B %d, %Y')
df
```

```python
df.isnull().sum()
```

Menyimpan data bersih di csv

```python
df_cl=pd.DataFrame(df)
df_cl.to_csv("df_cl.csv")
df_cl
```

Visualisasi

```python
df.type.value_counts()
```

```python
df.type.value_counts(normalize=True)
```

```python
plt.figure(figsize=(6,6))
plt.title("Number of Type Variable Values")
sns.set(style="darkgrid")
ax=sns.countplot(x="type", data=df, palette="Set2")
ax.bar_label(ax.containers[0])
```

The types of shows that are most frequently released on Amazon Prime are films compared to TV shows, namely films at 80.82% and TV shows at 19.18%.

```python
plt.figure(figsize=(30,25))
plt.title("Number of Ranking Variable Values")
sns.set(style="darkgrid")
ax=sns.countplot(x="rating", data=df, palette="Set2")
ax.bar_label(ax.containers[0])
```

The ranking shows that the most adds on Amazon Prime are from the 13+ level

```python
plt.figure(figsize=(12,5))
plt.title("Number of Categorical Variable Values")
sns.set(style="darkgrid")
ax = sns.countplot(x="categories", data=df, palette="Set2", order=df.categories.value_counts().iloc[:5].index)
ax.bar_label(ax.containers[0])
```

The drama category is the most frequently released program category on Amazon Prime.

```python
plt.figure(figsize=(12,2))
plt.title("Number of Country Variable Values")
sns.set(style="darkgrid")
ax = sns.countplot(x="country", data=df, palette="Set2", order=df.country.value_counts().iloc[:4].index)
ax.bar_label(ax.containers[0])
```

The top 3 countries that release the most TV shows and films on Amazon Prime are the United States, India and the United Kingdom.

```python
plt.figure(figsize=(12,3))
plt.title("Number of Director Variable Values")
sns.set(style="darkgrid")
ax = sns.countplot(x="director", data=df, palette="Set2", order=df.director.value_counts().iloc[:6].index)
ax.bar_label(ax.containers[0])
```

Overall, the data shows that a director named Mark Night is the director who has produced the most shows released on Amazon Prime, namely 113 shows.
